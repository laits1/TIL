```python
from requests import get
import requests
from bs4 import BeautifulSoup as Soup
import pandas as pd
import json
from tqdm.notebook import tqdm # show progress-bar
import re
import time
#크롬 드라이버를 통한  접속
from selenium.webdriver.chrome.options import Options 

from selenium import webdriver
from selenium.webdriver.common.keys import Keys

# 영화번호, [개인 user Id, 개인 user 평점 ], 영화 제목, 장르
```

```python
Name = []
Year = []
Rating = []
Meta_score = []
Votes = []
Genre = []
Movie_No = []
Director = []
Star = []
Actor_list = []

User_ID = []
User_rating_movie = []
User_rating = []
```

```python
movie = pd.read_csv("./Movie_No.csv", index_col=0)
```

```python
movie
```

```python
driver = webdriver.Chrome("./driver/chromedriver")
# driver.set_page_load_timeout("20")
```

```python
for i in tqdm(range(1)):      

    movieId = str(movie["Movie_No"][i+1])
    print(movieId)
    if len(movieId) == 6 :
        url = 'https://www.imdb.com/title/tt' + '0' + movieId + '/reviews?sort=submissionDate&dir=asc&ratingFilter=0'
        driver.get(url) # 영화 페이지
        time.sleep(2)
    else :
        url = 'https://www.imdb.com/title/tt' + movieId + '/reviews?sort=submissionDate&dir=asc&ratingFilter=0'
        driver.get(url) # 영화 페이지
        time.sleep(2)




    try:
        driver.find_element_by_class_name("ipl-load-more__button").send_keys(Keys.ENTER)
        time.sleep(2)
    except Exception as ex:
        print('더이상 "Load More" 없음', ex)
        

    print(" 댓글 수집 ")
    try:
        time.sleep(2)
        soup_data= Soup(driver.page_source, 'html.parser')
        movies_user_name = soup_data.findAll('div', {'class' : 'display-name-date'})
        movies_user_rating = soup_data.findAll('div', {'class' : 'lister-item-content'})

        time.sleep(1)
        
        for i in movies_user_name :
            User_ID.append(i.span.text)
            User_rating_movie.append(movieId)
        
        for i in range(len(movies_user_name)):
            
            try :
                if (movies_user_rating[i].span.text.strip()[-5:-3] == User_ID[i][-5:-3]) :
                    User_rating.append('0')
                else :
                    User_rating.append(movies_user_rating[i].span.text.strip()[-5:-3])
            except Exception as ex :
                print("뭔가 없음")
                print(movies_user_name[i].span.text)
                
    except Exception as ex:
        re=['nothing']
        print("댓글없음")

```

```python
data = list(zip(User_ID, User_rating_movie, User_rating))
df = pd.DataFrame(data, columns = ['User_ID', 'User_rating_movie','User_rating'])
```
