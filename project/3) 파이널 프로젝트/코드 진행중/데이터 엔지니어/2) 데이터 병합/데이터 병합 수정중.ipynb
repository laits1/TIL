### movies 기존데이터 + 2020~2021 데이터 + Comin_soon 데이터 병합

```python
movie = pd.read_csv('C:/Users/byungje/Desktop/마지막/손동기/ml-25m/ml-25m/movies.csv')
links = pd.read_csv('C:/Users/byungje/Desktop/마지막/손동기/ml-25m/ml-25m/links.csv')
imdb_movie = pd.merge(movie, links, on = 'movie_Id')
imdb_movie = imdb_movie[['imdbId', 'title', 'genres']]
```

```python
movie2000 = pd.read_csv('C:/Users/byungje/Desktop/마지막/2000-01~Current Top 1000 Movies/2000-01~Current Top 1000 Movies.csv')
```

```python
movie2000 = movie2000[['Movie_No','Name','Genre']]
movie2000 = movie2000.rename(columns = {'Movie_No':'imdbId', 'Name':'title', 'Genre':'genres'})
movie2000
```

```python
movie2020 = pd.read_csv('C:/Users/byungje/Desktop/마지막/2020-01~2021-09 Top 1000 Movies/2020_Movie_list.csv')
```

```python
movie2020 = movie2020[['Movie_No','Name','Genre', 'Director', 'Actor_list']]
movie2020 = movie2020.rename(columns = {'Movie_No':'imdbId', 'Name':'title', 'Genre':'genres'})
movie2020
```

```python
Comin_soon = pd.read_csv('C:/Users/byungje/Desktop/마지막/손동기/imdb Coming Soon.csv')
```

```python
Comin_soon = Comin_soon[['MovieId', 'Name', 'Genre', 'Director', 'Stars']]
Comin_soon = Comin_soon.rename(columns={'MovieId':'imdbId', 'Name':'title', 'Genre':'genres', 'Stars' : 'Actor_list'})
Comin_soon
```

```python
Total.to_csv('movies.csv')
```

### Director 데이터
    - 감독 목록 추출 후 감독 테이블 생성
    - 개개인의 감독마다 번호 부여

```python
Director = Total[['Director_name']]
Director = Director.dropna()
Director = Director.reset_index()
Director = Director.drop(['imdb_movie_Id'], axis = 1)
Director = Director.reset_index()
Director = Director.rename(columns = {'index' : 'Director_Id'})
Director = Director.set_index('Director_Id')
Director
```

```python
Director.to_csv('Director.csv')
```

### Actor 데이터
    - 배우 목록 추출 후 감독 테이블 생성
    - 개개인의 배우마다 번호 부여

```python
b = list(Total['Actor_name'].dropna())
```

```python
actor = []
for i in range(len(b)) :
    for j in range(len(b[i][1:-1].split(','))) :
        if j == 0 :
            actor.append(b[i][1:-1].split(',')[j][1:-1])
        else :
            actor.append(b[i][1:-1].split(',')[j][2:-1])
```

```python
df_actor = pd.DataFrame(actor)
df_actor
```

```python
df_actor = df_actor.drop_duplicates()
df_actor = df_actor.reset_index()
df_actor = df_actor.reset_index()
df_actor = df_actor.rename(columns = {0 : 'Actor_name', 'level_0' : 'Actor_Id'})
df_actor = df_actor.drop('index', axis = 1)
df_actor = df_actor.set_index('Actor_Id')
df_actor
```

```python
df_actor.to_csv('Actor.csv')
```
