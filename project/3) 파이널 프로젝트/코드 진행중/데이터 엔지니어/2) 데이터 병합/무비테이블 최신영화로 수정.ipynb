```python
import pandas as pd
```

```python
movies = pd.read_csv('./after_csv/movies.csv', usecols=['imdb_movie_Id','title','genres'])
```

```python
for i in range(len(movies)) :
    try :
        print(movies['title'][i][-5:-1])
        if(int(movies['title'][i][-5:-1]) < 2000) :
            movies = movies.drop([i])
    except :
        pass
```

```python
movies = movies.set_index('imdb_movie_Id')
```
