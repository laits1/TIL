```python
import pandas as pd
from requests import get
import requests
from bs4 import BeautifulSoup as Soup
from tqdm import tqdm 
import time
```

```python
movies = pd.read_csv('./movies.csv', usecols = ['imdb_movie_Id','title'])
```

```python
movies
```

```python
images = []
for i in tqdm(range(len(movies))) :
    try :
        if i == 5486 :
            break
        url = get('https://www.imdb.com/title/tt' + format(movies.imdb_movie_Id[i+5488], '07'))
        req = url.text
        soup_data = Soup(req, 'html.parser')
    
        image_tag = soup_data.findAll('div', {'class' : 'ipc-poster ipc-poster--baseAlt ipc-poster--dynamic-width Poster__CelPoster-sc-6zpm25-0 kPdBKI celwidget ipc-sub-grid-item ipc-sub-grid-item--span-2'})
        images.append(image_tag[0].div.img['srcset'].split(',')[-4])
    except :
        print('이미지 없음', i+5488)
        image_tag = 'https://i.pinimg.com/originals/ca/8a/69/ca8a69672befd7b741d703de05bf0d0b.png'
        images.append(image_tag)
```

```python
movies['image'] = 1
```

```python
for i in tqdm(range(len(movies))) :
    if i == 5486 :
        break
        
    movies['image'][i+5488] = images[i]
```
