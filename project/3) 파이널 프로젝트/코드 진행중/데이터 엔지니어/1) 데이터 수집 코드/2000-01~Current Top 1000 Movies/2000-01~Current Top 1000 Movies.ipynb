```python
from requests import get
from bs4 import BeautifulSoup as Soup
import pandas as pd
import time
```

```python
Name = []
Year = []
Rating = []
Meta_score = []
Votes = []
Genre = []
Movie_No = []
Director = []
Star = []
Actor_list = []
```

```python
for n in range (10) :
    url = get('https://www.imdb.com/search/title/?title_type=feature,tv_movie,short&release_date=2000-01-01,&count=100&start=' + str(n*100+1) + '&ref_=adv_nxt')
    request = url.text
    soup_data = Soup(request, 'html.parser')
    movies = soup_data.findAll('div', {'class' : 'lister-item mode-advanced'})
    for i in movies:
        Name.append(i.h3.a.text)
        Year.append(i.find('span', {'class' : 'lister-item-year text-muted unbold'}).text[-5:-1])
        try :    
            Rating.append(i.find('div', {'class' : 'inline-block ratings-imdb-rating'})['data-value'])
        except:
            Rating.append('0')
        Genre.append(i.find('span', {'class' : 'genre'}).text[1:-1])
        Director.append(i.select("div.lister-item-content> p:nth-of-type(3) a")[0].text)
        Movie_No.append(i.h3.a.get('href').split('/')[2][2:])
        Actor_list.append(i.select("div.lister-item-content> p:nth-of-type(3) a")[1:])
        try :    
            Votes.append(i.find('span', {'name' : 'nv'})['data-value'])
        except :
            Votes.append(0)
        try :
            Meta_score.append(i.find('div', {'class' : 'inline-block ratings-metascore'}).span.text.strip())
        except :
            Meta_score.append(0)

for k in range(len(Actor_list)):
    for j in range(len(Actor_list[k])):
        Actor_list[k][j] = Actor_list[k][j].text
            
            
print(len(Name))
print(len(Rating))
print(len(Actor_list))
```

```python
data = list(zip(Name, Year, Genre, Rating, Meta_score, Votes, Movie_No, Director, Actor_list))
```

```python
df = pd.DataFrame(data, columns = ['Name', 'Year','Genre', 'Rating', 'Metascore',  'Votes', 'Movie_No', 'Director', 'Actor_list'])
df.index = df.index+1
```
