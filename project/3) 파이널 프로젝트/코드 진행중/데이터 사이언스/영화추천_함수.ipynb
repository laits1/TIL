```python
import pandas as pd # 데이터 핸들링 
import numpy as np 
import warnings
import json
from ast import literal_eval
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
warnings.filterwarnings("ignore")
```

```python
genre_sim= pd.read_csv('./data/data.csv')
```

```python
genre_sim.set_index('title')
```

```python
# 장르가 유사한 영화 추천 함수 선언

def REC(title, genre_sim= genre_sim):
    choice= []
    
    # 모든 영화에 대한 해당 영화의 유사도 계산
    sim_scores= list(enumerate(genre_sim[title]))
    
    # 유사도에 따라 영화 정렬
    sim_scores= sorted(sim_scores, key= lambda x: x[1], reverse= True)
    
    # 가장 유사한 10개의 영화 받아옴
    sim_scores= sim_scores[1:11]
    
    # 가장 유사한 10개 영화의 인덱스를 추출
    movie_indices= [i[0] for i in sim_scores]
    
    return genre_sim['title'].iloc[movie_indices]
```
