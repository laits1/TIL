```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"
from pandas_datareader import data
import matplotlib.pyplot as plt
from datetime import datetime

import pandas as pd
import numpy as np
```

```python
def graph(corp):
    spy = pd.read_csv('./price/spy_price.csv')
    spy.index=pd.to_datetime(spy['date'])
    stock = data.DataReader([corp],'yahoo','2019-08-02','2021-08-31')
    stock['Adj Close']
    
    fig, ax1 = plt.subplots(figsize=(10,6))
    ax2 = ax1.twinx()

    ax1.set_title(corp+' vs SPY', fontsize=15)
    ax1.set_ylabel(corp)
    ax2.set_ylabel('spy')

    ax1.set_xlabel("Date Time")

    ax1.plot(stock['Adj Close'],color='blue')
    ax2.plot(spy['Adj Close'],color='red')

    ax1.legend([corp+' ($)'],loc='upper left')
    ax2.legend(['spy'],loc='upper right')
    plt.show()
```

Corp : 'TSLA'  'MSFT'  'AMZN'  'SPOT'  'PFE'  'AMD'   'WMT'  'MO'  'KO'    'NKE'   'FB'

```python
graph('PFE')
```
