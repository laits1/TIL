```python
TSLA = pd.read_csv('./stock_price_2y.csv', 
                         index_col = 'Date',
                        usecols=['Date','TSLA'])

MSFT = pd.read_csv('./stock_price_2y.csv', 
                         index_col = 'Date',
                        usecols=['Date','MSFT'])

AMZN = pd.read_csv('./stock_price_2y.csv', 
                         index_col = 'Date',
                        usecols=['Date','AMZN'])
```

```python
ETF_ = pd.read_csv('./ETF_price_2y.csv', 
                         index_col = 'Date',
                        )
```

```python
TSLA['IYK'] = ETF_['IYK']
MSFT['IXN'] = ETF_['IXN']
AMZN['IBUY'] = ETF_['IBUY']
TSLA.plot(figsize=(12,9))
MSFT.plot(figsize=(12,9))
AMZN.plot(figsize=(12,9))
```

```python
TSLA['Date'] = TSLA.index
TSLA['Date']

TSLA['Date'] = pd.to_datetime(TSLA['Date'])
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
fig, axe1 = plt.subplots()
axe2 = axe1.twinx()
c1 = sns.lineplot(ax = axe1, data = TSLA, x = TSLA['Date'], y = 'TSLA', color = 'red')
c2 = sns.lineplot(ax = axe2, data = TSLA, x = TSLA['Date'], y = 'IYK', color = 'blue')

axe1.legend(['TSLA', 'IYK'])

axe1.set_ylabel('TSLA')
axe2.set_ylabel('IYK')


# fig
```
