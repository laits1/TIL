### TESLA 엑셀 불러오기

```python
tesla = pd.read_excel('TESLA.xlsx')
tesla
```

### url을 리스트화 하여 for문에 사용하기 위해 DF 변경

```python
tesla = tesla.rename(columns=tesla.iloc[0])
tesla.drop(index=0,inplace=True)
tesla = tesla[['TESLA']]
```

```python
tesla # 2번이 NONE이므로 밑에서 확인 할 것
```

```python
tesla_lists = tesla['TESLA'].tolist() # 1~101 기사를 리스트화 시켜 for문에 이용하기 위함
tesla_lists
```

### 모든기사 크롤링 자동화 작업

```python
driver = webdriver.Chrome(executable_path='../driver/chromedriver.exe')

tesla_article=[] # 기사를 넣을 빈 리스트 작성
  
for i in tqdm(tesla_lists[0:3]):       # 예시로 1~3번까지 (2번 NONE 확인 차)

    if i == "NONE":
        tesla_article.append("NONE")   # 만약 NONE이면 기사대신 NONE으로 입력
    else:
        driver.get(i)                  # i = 기사의 url
        html = driver.page_source
        soup = bs4.BeautifulSoup(html, 'html.parser')
        
        div = soup.find('div',{'class':'caas-body'})   # 기사 내용이 div class='caas-body'에 들어있음 (하이퍼링크내용 제거는 차후 코딩...)
        article = div.text                             # text화 시켜 하나의 내용으로 변경
        
        tesla_article.append(article)                  # 기사를 넣을 빈 리스트에 차례대로 append
        
driver.close() # 크롬 닫기
driver.quit()  # 크롬 끄기 (메모리 차지 심해서 끝나고 모두 끄기)
```

```python
tesla_article # 2번 NONE 리스트에 NONE으로 들어간 걸 확인
```
