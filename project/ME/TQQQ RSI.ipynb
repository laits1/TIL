RSI 30이하 = 졸업 (주식100%)

RSI 30~35 = 2차 매수 고려 (주식80%)

RSI 40~45 = 1차 매수 고려 (주식60%)



RSI 50은 중립



RSI 60으로 오르면, 1차 매도 고려 (주식50%)

RSI 70으로 오르면, 2차 매도 고려 (주식25%)

RSI 80으로 오르면, 졸업 (주식0%)



```python
from pandas_datareader import data
from datetime import datetime
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def message(i) :
    print(f"==={i}일 째===")
    print(f"RSI         : {RSI.RSI[i]}")
    print(f"balance     : {balance}")
    print(f"{Ticker} 현재가 : {RSI['Adj Close'][i]}")
    print(f"qty         : {qty}")
    print(f"총 자산     : {balance + (RSI['Adj Close'][i] * qty)}")
    print("")
```

```python
def cal(Ticker, start, end, balance) :
    
    
    global ticker 
    ticker = data.DataReader(Ticker,'yahoo', start, end)
    global delta 
    delta = ticker['Close'].diff()
    global up 
    up = delta.clip(lower=0)
    global down 
    down = -1*delta.clip(upper=0)
    global ema_up 
    ema_up = up.ewm(com=13, adjust=False).mean()
    global ema_down 
    ema_down = down.ewm(com=13, adjust=False).mean()
    global rs 
    rs = ema_up/ema_down

    ticker['RSI'] = 100 - (100/(1 + rs))

    # Skip first 14 days to have real values
    ticker = ticker.iloc[14:]
    global RSI 
    RSI= ticker[['Adj Close','RSI']]

    
    print(f"===0일 째===")
    print(f"RSI         : 0")
    print(f"balance     : {balance}")
    print(f"{Ticker} 현재가 : 0")
    print(f"qty         : 0")
    print(f"총 자산     : {balance}")
    print("")

    global total_yield 
    total_yield = []
    global qty
    qty = 0
   
    
    
    for i in range(len(RSI)) :
        total_yield.append(balance + (RSI['Adj Close'][i] * qty))
        
        if (RSI.RSI[i] <= 30) :
            qty += divmod(balance,RSI['Adj Close'][i])[0]
            print(balance - divmod(balance,RSI['Adj Close'][i])[0] * RSI['Adj Close'][i])
            balance = balance - (divmod(balance,RSI['Adj Close'][i])[0] * RSI['Adj Close'][i])

        elif (30 < RSI.RSI[i] <= 35) :
            qty += divmod((balance * 0.5) , RSI['Adj Close'][i])[0]
            balance += - (divmod((balance * 0.6) , RSI['Adj Close'][i])[0] * RSI['Adj Close'][i])

        elif (35 < RSI.RSI[i] <= 50) :
            qty += divmod((balance * 0.6) , RSI['Adj Close'][i])[0]
            balance += - (divmod((balance * 0.6) , RSI['Adj Close'][i])[0] * RSI['Adj Close'][i])

        elif (50 < RSI.RSI[i] <= 60) :
            pass
        
        elif (60 <= RSI.RSI[i] < 70) :
            qty += - divmod(qty,2)[0]
            balance += qty * RSI['Adj Close'][i]

        elif (70 <= RSI.RSI[i] < 80) :
            qty += - divmod(qty,2)[0]
            balance += qty * RSI['Adj Close'][i]

        elif (80 <= RSI.RSI[i] < 101) :
            balance += qty * RSI['Adj Close'][i]
            qty =0

        print(f"==={i}일 째===")
        print(f"RSI         : {RSI.RSI[i]}")
        print(f"balance     : {balance}")
        print(f"{Ticker} 현재가 : {RSI['Adj Close'][i]}")
        print(f"qty         : {qty}")
        print(f"총 자산     : {balance + (RSI['Adj Close'][i] * qty)}")
        print("")
```
