### 사용자 정의 함수를 이용해 ','를 없애고 int형으로 변환해 barh 그래프로 표시!

### thousands -> 질문! (안된다!)
### .astype(int) -> ','가 있어서 사용 불가능

```python
# 사용자 정의 함수!
# ','를 없애고, int형으로 바꾸기!

def to_int(data) :
    data = data.replace(',','')
    data = int(data)
    return data
```

```python
# 위의 함수를 이용해 ,를 삭제하고 int형으로 바꾸기

hospital_2017_int = hospital_2017.계.apply(to_int)
hospital_2018_int = hospital_2018.계.apply(to_int)
hospital_2019_int = hospital_2019.계.apply(to_int)
hospital_2020_int = hospital_2020.계.apply(to_int)
```

```python
# 만든 Series를 첫번째 열에 삽입

hospital_2017.insert(1, '합계_int', hospital_2017_int)
hospital_2018.insert(1, '합계_int', hospital_2018_int)
hospital_2019.insert(1, '합계_int', hospital_2019_int)
hospital_2020.insert(1, '합계_int', hospital_2020_int)
```

```python
# 원래의 "합계_발생" 열은 삭제!

hospital_2017.drop(columns=["계"],inplace=True)
hospital_2018.drop(columns=["계"],inplace=True)
hospital_2019.drop(columns=["계"],inplace=True)
hospital_2020.drop(columns=["계"],inplace=True)
```

```python
gu_name = list(hospital_fin_2017['자치구'])
```

- geocoder
    - 고유명칭(주소, 산, 호수의 이름 등)을 가지고 위도(lat)와 경도(lng)의 좌표 값 등 정보를 얻어오는 기능
    - geocode(고유명칭, 언어방식)

```python
# 빈 리스트 3개를 만든다 (주소, 위도, 경도)
gu_address=[]
gu_lat=[]
gu_lng=[]
```

```python
for name in gu_name :
    tmp = gmaps.geocode(name,language='ko') # geocode(고유명칭, 언어방식)
    gu_address.append(tmp[0].get('formatted_address')) # 구 주소를 저장 : dict.get(key) -> key에 대한 value 반환
    
    # 위/경도 추출
    tmp_loc = tmp[0].get('geometry')
    
    gu_lat.append(tmp_loc['location']['lat']) # latitude 속성이 위도를, 
    gu_lng.append(tmp_loc['location']['lng']) # longitude 속성이 경도를  나타낸다! / 여러 번 실행하면 계속 추가되므로 한번만!
    
    print(name + '------->' + tmp[0].get('formatted_address'))  
```

# 2017년도 시각화

```python
# 서울시 중심부의 위도, 경도 입니다.
center = [37.541, 126.986]

# 맵이 center 에 위치하고, zoom 레벨은 11로 시작하는 맵을 만듭니다.
map17 = folium.Map(location=center,tiles='cartodbpositron',  zoom_start=11)

# Choropleth 레이어를 만들고, 맵에 추가합니다.
folium.Choropleth(
    geo_data=geo_data, # 지도 데이터 파일 경로 (.geojson, geopandas.DataFrame)
    data=hospital_fin_2017, # 시각화 하고자 하는 데이터파일 (pandas.DataFrame)
    columns=('자치구', '의료기관_총_합계'), # (지도 데이터와 매핑할 값, 시각화 하고자하는 변수)
    key_on='feature.properties.name', #  feature.데이터 파일과 매핑할 값
    fill_color='Paired', # 시각화에 쓰일 색상 (Paired)
    legend_name='서울시 각 구별 의료기관 총 합계 (2017)', # 범주 이름

).add_to(map17)

# 맵 map을 출력합니다.
map17
```
