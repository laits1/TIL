```python
crime_data_17 = pd.read_csv("./data/서울시_5대_범죄_발생건수(2017)_최종.csv",thousands = ',') 
crime_data_18 = pd.read_csv("./data/서울시_5대_범죄_발생건수(2018)_최종.csv",thousands = ',')
crime_data_19 = pd.read_csv("./data/서울시_5대_범죄_발생건수(2019)_최종.csv",thousands = ',') 
```

```python
population_data_17 = pd.read_csv("./data/2017년_연령별_인구현황.csv",thousands = ',') 
population_data_18 = pd.read_csv("./data/2018년_연령별_인구현황.csv",thousands = ',')
population_data_19 = pd.read_csv("./data/2019년_연령별_인구현황.csv",thousands = ',') 
```

```python
crime_data_17.head(3)
population_data_17.head(3)
```

```python
crime_data_17["10만명 당 범죄 발생 건수"] = crime_data_17["범죄_발생_합계"] / population_data_17["총인구수"] * 100000
crime_data_18["10만명 당 범죄 발생 건수"] = crime_data_17["범죄_발생_합계"] / population_data_18["총인구수"] * 100000
crime_data_19["10만명 당 범죄 발생 건수"] = crime_data_17["범죄_발생_합계"] / population_data_19["총인구수"] * 100000
```

```python
hospital_data_17 = pd.read_csv("./data/서울시_의료기관_구_별_합계(2017)_최종_종합병원_수_포함.csv",thousands = ',') 
hospital_data_18 = pd.read_csv("./data/서울시_의료기관_구_별_합계(2018)_최종_종합병원_수_포함.csv",thousands = ',') 
hospital_data_19 = pd.read_csv("./data/서울시_의료기관_구_별_합계(2019)_최종_종합병원_수_포함.csv",thousands = ',') 
hospital_data_20 = pd.read_csv("./data/서울시_의료기관_구_별_합계(2020)_최종_종합병원_수_포함.csv",thousands = ',')
```

```python
hospital_data_17.head(3)
hospital_data_18.head(3)
hospital_data_19.head(3)
hospital_data_20.head(3)
```

```python
hospital_data_17["10만명 당 의료기관 수"] = hospital_data_17["의료기관_총_합계"] / population_data_17["총인구수"] * 100000 # 10만 명당
hospital_data_18["10만명 당 의료기관 수"] = hospital_data_18["의료기관_총_합계"] / population_data_18["총인구수"] * 100000
hospital_data_19["10만명 당 의료기관 수"] = hospital_data_19["의료기관_총_합계"] / population_data_19["총인구수"] * 100000
hospital_data_20["10만명 당 의료기관 수"] = hospital_data_20["의료기관_총_합계"] / population_data_19["총인구수"] * 100000
```

```python
hospital_data_17["10만명 당 종합병원 수"] = hospital_data_17["종합병원"] / population_data_17["총인구수"] * 100000 # 10만 명당
hospital_data_18["10만명 당 종합병원 수"] = hospital_data_18["종합병원"] / population_data_18["총인구수"] * 100000
hospital_data_19["10만명 당 종합병원 수"] = hospital_data_19["종합병원"] / population_data_19["총인구수"] * 100000
hospital_data_20["10만명 당 종합병원 수"] = hospital_data_20["종합병원"] / population_data_19["총인구수"] * 100000
```
