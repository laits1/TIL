```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_ex = pd.read_csv('환율통계.csv')
df_ex
```

```python
df_loan = pd.read_csv('대출통계.csv')
df_loan 
```

```python
df_kospi = pd.read_csv('연도별코스피.csv')
df_kospi.rename(columns = {'년도': '년월'}, inplace = True)
```

```python
df_apt = pd.read_csv('부동산월별가격.csv')
df_apt.rename(columns = {'계약날짜': '년월'}, inplace = True)
df_apt
```

```python
df5 = pd.read_csv('가계대출.csv')
df6 = pd.read_csv('월별금리.csv')
df7 = pd.read_csv('주택대출.csv')
df7
```

```python
df_eco = pd.concat([df_ex,df_kospi,df6,df5,df7,df_apt],ignore_index=True) ## 각각의 데이터 concat
df_eco
```

```python
dfEco = df_eco.transpose()	#행 열 전환
dfEco.rename(columns=dfEco.iloc[0], inplace=True)	# 행열이 전환된 데이터프레임의 열 이름 제대로 수정
dfEco = dfEco.drop(dfEco.index[0])
dfEco.rename(columns = {'서울시 가계대출량(10억원)': '가계대출(10억원)'}, inplace = True)
dfEco
```
