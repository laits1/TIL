    txt파일을 csv로 불러오기 위해서 delimiter '\t' 코드를 삽입해 주었다. 각 컬럼을 \t로 구분해 파일을 불러온다는 뜻인것 같다.
    합계, 소계등 필요 없는 칼럼은 삭제 해주어야 할 필요가 있다.

```python
populationDensity = pd.read_csv('./data/populationDensity.txt',delimiter = '\t')
```

```python
populationDensity
```

```python
populationDensity[populationDensity["동"]=='합계']
```

```python
populationDensity.drop(0,inplace=True)
```

```python
populationDensity
```

```python
populationDensity.drop(populationDensity[populationDensity["동"]=="소계"].index,inplace=True)
```

```python
# 인덱스를 재설정해주어야한다.
populationDensity["index"]=range(0,425)
```

```python
populationDensity
```

```python
populationDensity.set_index("index",inplace=True)
```

```python
populationDensity
```

```python
populationDensity[populationDensity["동"]=="무악동"]["인구밀도"]
```

```python
populationDensity["면적"]=populationDensity["면적"].astype("float")
populationDensity["인구"]=populationDensity["인구"].str.replace(pat=r'[^\w]', repl=r'', regex=True)
```

```python
populationDensity["인구"]=populationDensity["인구"].astype("float")
```

```python
populationDensity["인구"]
```

```python
seoul_gu=["노원구",
"송파구",
"강서구",
"강남구",
"강동구",
"구로구",
"성북구",
"양천구",
"도봉구",
"서초구",
"영등포구",
"성동구",
"동대문구",
"마포구",
"동작구",
"은평구",
"중랑구",
"관악구",
"서대문구",
"강북구",
"용산구",
"광진구",
"금천구",
"중구",
"종로구",]
```

```python
density=[]
for gu in seoul_gu:
    density.append(populationDensity[populationDensity["자치구"]==gu]["인구"].sum()/populationDensity[populationDensity["자치구"]==gu]["면적"].sum())
```

```python
density
```

```python
seoul_density=pd.DataFrame([seoul_gu,density]).T
seoul_density.columns=["구","인구밀도"]
seoul_density.sort_values(by="구", ascending=True, inplace=True)
```

```python
seoul_density
```

```python
seoul_density.to_csv('./fin_data/2020년서울시인구밀도.csv',header=True,index=False)
```
