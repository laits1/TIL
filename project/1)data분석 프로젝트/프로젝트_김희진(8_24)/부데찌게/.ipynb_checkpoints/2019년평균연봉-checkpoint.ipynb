## read_csv 하는데 필요한 파일 목록
 - ./data/2019연말정산.csv

```python
state19 = pd.read_csv("./data/2019연말정산.csv", encoding='CP949')
```

```python
state19.head(30)
```

```python
# 2019 연말장소 파일에서 서울시 구별로 인구와 총 급여만 나타낸 데이터 프레임을 추출한다
state19=state19.iloc[2:27,1:4]
```

```python
state19
```

```python
state19["인원 (명)"]
```

```python
state19["금액 (백만원)"]
```

```python
# 서울시 구별 평균 연봉을 알기 위해서 구별 총 급여를 구별 인구로 나누어 주었다.
state19["평균연봉(원)"]=state19["금액 (백만원)"]/state19["인원 (명)"]*10000
```

```python
# 인덱스 리셋
state19.reset_index(drop=True,inplace=True)
```

```python
# 구 명과 구별 평균연봉만 칼럼으로 추출하였다.
seoulSalary19=state19[["행정구역(시군구)별(2)","평균연봉(원)"]]
```

```python
seoulSalary19
```

```python
seoulSalary19.to_csv('./fin_data/2019서울시구별평균연봉.csv',header=True,index=False)
```
