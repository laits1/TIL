```python
apart17 = pd.read_csv("./data/2017아파트실거래가.csv", encoding='CP949')
apart18 = pd.read_csv("./data/2018아파트실거래가.csv", encoding='CP949')
apart19 = pd.read_csv("./data/2019아파트실거래가.csv", encoding='CP949')
apart20 = pd.read_csv("./data/2020아파트실거래가.csv", encoding='CP949')
```

```python
pd.concat([apart17,apart18,apart19,apart20])
```

```python
# 구 명을 데이터 프레임으로 만들때 사용하기 위해 따로 리스트를 작성해두었다. 
seoul_gu=["노원구",
"송파구",
"강서구",
"강남구",
"강동구",
"구로구",
"성북구",
"양천구",
"도봉구",
"서초구",
"영등포구",
"성동구",
"동대문구",
"마포구",
"동작구",
"은평구",
"중랑구",
"관악구",
"서대문구",
"강북구",
"용산구",
"광진구",
"금천구",
"중구",
"종로구",]
```

```python
for i in range(17,21):
    # 주택 평균 가격을 담을 리스트를 추가하였다. 
    globals()['medv'+str(i)] = []
    # 거래금액이 문자형으로 저장되어있고, 자리수 기호(,) 때문에 정수형으로 바꿔지지 않아서 자리수 기호르 삭제해주었다.
    globals()['apart'+str(i)]["거래금액(만원)"]=globals()['apart'+str(i)]["거래금액(만원)"].str.replace(pat=r'[^\w]', repl=r'', regex=True)
    globals()['apart'+str(i)]["거래금액(만원)"].head(3)
    # 거래금액을 수치형으로 변환해주었다.
    globals()['apart'+str(i)]["거래금액(만원)"]=globals()['apart'+str(i)]["거래금액(만원)"].astype("float")
    globals()['apart'+str(i)]["거래금액(만원)"].head(3)
    # 구 별 자료를 쓰기 위해서 주소를 기준으로 시,구,동으로 나누어주었다. (이 부분도 원래는 "군"을 고치지 않았는데 더 정확한 정보를 표기 하도록 수정)
    globals()['apart'+str(i)][['시','구','동']] = globals()['apart'+str(i)].시군구.str.split(" ",expand=True)
    globals()['apart'+str(i)]["거래금액(만원)/전용면적(㎡)"]=globals()['apart'+str(i)]["거래금액(만원)"]/globals()['apart'+str(i)]["전용면적(㎡)"]

    for gu in seoul_gu:
        globals()['medv'+str(i)].append(globals()['apart'+str(i)][globals()['apart'+str(i)]["구"]==gu]["거래금액(만원)/전용면적(㎡)"].mean())
    globals()['df'+str(i)]=pd.DataFrame([seoul_gu,globals()['medv'+str(i)]]).T
    globals()['df'+str(i)].columns=["구","거래금액(만원)/전용면적(㎡)"]
    # 자료를 처음 보는 사람들도 년도를 빨리 파악할 수 있게 날짜를 추가해달라는 피드백을 참고하여 날짜 칼럼을 추가하였다(great)
    globals()['df'+str(i)]["년도"]="20"+str(i)
    # 1평은 약 3.3 제곱이므로 3.3을 곱하여 1평당 아파트 평균 가격을 나타내줄 칼럼을 추가하였다. 
    globals()['df'+str(i)]["거래금액(만원)/1평"]=globals()['df'+str(i)]["거래금액(만원)/전용면적(㎡)"]*3.305785
    # 많이 쓰이는 국민 평수인 32평당 아파트 평균 가격을 나타내줄 칼럼을 추가하였다. 
    globals()['df'+str(i)]["거래금액(만원)/국민평수(32평)"]=globals()['df'+str(i)]["거래금액(만원)/전용면적(㎡)"]*3.305785*32
    globals()['df'+str(i)]=globals()['df'+str(i)][["구","년도", "거래금액(만원)/전용면적(㎡)", "거래금액(만원)/1평", "거래금액(만원)/국민평수(32평)"]]
    globals()['df'+str(i)].sort_values(by="구", ascending=True, inplace=True)
    globals()['df'+str(i)].reset_index(drop=True,inplace=True)
    globals()['df'+str(i)].to_csv('./fin_data/20'+str(i)+'년도_서울시_아파트_평균금액.csv',header=True,index=False)
```
