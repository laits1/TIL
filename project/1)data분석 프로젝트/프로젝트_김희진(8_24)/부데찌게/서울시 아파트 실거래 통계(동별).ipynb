```python
apart17 = pd.read_csv("./data/2017아파트실거래가.csv", encoding='CP949')
apart18 = pd.read_csv("./data/2018아파트실거래가.csv", encoding='CP949')
apart19 = pd.read_csv("./data/2019아파트실거래가.csv", encoding='CP949')
apart20 = pd.read_csv("./data/2020아파트실거래가.csv", encoding='CP949')
```

```python
apart7890=pd.concat([apart17,apart18,apart19,apart20])
```

```python
apart7890[['시','구','동']] = apart7890.시군구.str.split(" ",expand=True)
```

```python
for i in range(7890,7891):
    # 주택 평균 가격을 담을 리스트를 추가하였다. 
    globals()['medv'+str(i)] = []
    # 거래금액이 문자형으로 저장되어있고, 자리수 기호(,) 때문에 정수형으로 바꿔지지 않아서 자리수 기호르 삭제해주었다.
    globals()['apart'+str(i)]["거래금액(만원)"]=globals()['apart'+str(i)]["거래금액(만원)"].str.replace(pat=r'[^\w]', repl=r'', regex=True)
    globals()['apart'+str(i)]["거래금액(만원)"].head(3)
    # 거래금액을 수치형으로 변환해주었다.
    globals()['apart'+str(i)]["거래금액(만원)"]=globals()['apart'+str(i)]["거래금액(만원)"].astype("float")
    globals()['apart'+str(i)]["거래금액(만원)"].head(3)
    # 구 별 자료를 쓰기 위해서 주소를 기준으로 시,구,동으로 나누어주었다. (이 부분도 원래는 "군"을 고치지 않았는데 더 정확한 정보를 표기 하도록 수정)
    globals()['apart'+str(i)][['시','구','동']] = globals()['apart'+str(i)].시군구.str.split(" ",expand=True)
    globals()['apart'+str(i)]["거래금액(만원)/전용면적(㎡)"]=globals()['apart'+str(i)]["거래금액(만원)"]/globals()['apart'+str(i)]["전용면적(㎡)"]
```
